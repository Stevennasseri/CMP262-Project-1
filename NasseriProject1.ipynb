### I see two reasonable options to take care of this issue. it seems as though some of the instances had multiple options selected. Option 1 would be to combine all instances of multiple entries into one 'mixed' category. option 2 would be to separate the selected races, and credit each category once. I'm going to do option 1, because I feel like the numbers will be more accurate and helpful in calculations

```python
## I used chatgpt to help me make the first function

def update_race(entry): # function to update each instance in a specific column
    if ";" in entry:  # searches for any instances with a semicolon, indicating multiple entries were selected
        return "mixed"  # returns a new value to the instance to create a new 'mixed' category
    else:
        return entry  # for every other instance with only one entry, return the initial value

# Apply the function to the 'Race/ethnicity' column
df1['Race/ethnicity'] = df1['Race/ethnicity'].apply(update_race)  # apply function iterates over every instance in the selected column

df2['Race/ethnicity'] = df2['Race/ethnicity'].apply(update_race)


# use the same idea to iterate over each instance and replace spaces with _
def remove_spaces(entry):
    if ' ' in entry:  # if entry has a space in it
        return entry.replace(' ', '_')  # replace space with underscore
    else:
        return entry  # if no space, return initial value
        
df1['Race/ethnicity'] = df1['Race/ethnicity'].apply(remove_spaces)

df1['Race/ethnicity'].value_counts()  # view
```

### Starting with data file 1, we can combine all instancing including common words

```python
def major_update(entry):  # function to narrow down possible majors
    if 'engineering' in entry.lower():  # engineering
        return 'engineering'
    elif 'business' in entry.lower():  # business
        return 'business'
    elif 'bio' in entry.lower():  # biology
        return 'biology'
    elif 'computer science' in entry.lower():  # comp sci
        return 'computer science'
    elif 'education' in entry.lower():  # education
        return 'education'
    elif 'information' in entry.lower():  # IT
        return 
    elif 'highschool' in entry.lower():  # high school students
        return 'highschool'
    elif 'certificate' in entry.lower():  # certificate programs
        return 'certificate program'
    elif 'edam' in entry.lower():  # EDAM
        return 'edam'
    elif 'psychology' in entry.lower():
        return 'psychology'
    elif 'radiography' in entry.lower():
        return 'radiography'
    else:
        return entry
    
# apply and view update in df1
df1['current_program'] = df1['current_program'].apply(major_update)
df1['current_program'].value_counts()
```

```python
# apply and view update in df2
df2['current_program'] = df2['current_program'].apply(major_update)
df2['current_program'].value_counts()
```
